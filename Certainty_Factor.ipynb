```python
# Gejala user dengan nilai CF awal
gejala_user = {
    "demam": 0.9,
    "batuk": 0.6,
    "sakit_tenggorokan": 0.6,

#menambahkan 5 gejala baru beserta nilai CF nya.
    "pilek": 0.5,
    "sakit_kepala": 0.4,
    "lelah": 0.5,
    "bersin": 0.3,
    "pegal": 0.4
}

# Pengetahuan pakar tentang gejala flu
pengetahuan = {
    "flu": {
        "demam": 0.8,
        "batuk": 0.7,
        "sakit_tenggorokan": 0.6,
        "pilek": 0.7,
        "sakit_kepala": 0.5,
        "lelah": 0.6,
        "bersin": 0.4,
        "pegal": 0.5
    }
}

# Fungsi Certainty Factor
def hitung_cf(gejala_user, pengetahuan_pakar):
    cf_total = 0
    first = True
    for gejala, cf_user in gejala_user.items():
        if gejala in pengetahuan_pakar:
            cf_pakar = pengetahuan_pakar[gejala]
            cf = cf_user * cf_pakar
            if first:
                cf_total = cf
                first = False
            else:
                cf_total = cf_total + cf * (1 - cf_total)
    return cf_total

# Hitung CF awal
cf_awal = hitung_cf(gejala_user, pengetahuan["flu"])
print(f"CF diagnosis Flu (sebelum perubahan): {cf_awal:.2f}")

# Ubah nilai demam menjadi 0.2
gejala_user["demam"] = 0.2

# Hitung CF setelah perubahan
cf_setelah = hitung_cf(gejala_user, pengetahuan["flu"])
print(f"CF diagnosis Flu (setelah demam diubah jadi 0.2): {cf_setelah:.2f}")
```

Hasil diagnosis flu pada sistem pakar ini dipengaruhi oleh penurunan nilai CF gejala "demam" dari 0,9 menjadi 0,2, yang memiliki bobot pakar tinggi (0,8). Awalnya, kontribusi "demam" signifikan (0,72), namun setelah perubahan, kontribusinya turun drastis menjadi 0,16. Meskipun gejala lain seperti "pilek" dan "pegal" ada, kontribusinya tidak cukup besar untuk mengimbangi penurunan pada "demam". Hal ini menunjukkan bahwa dalam sistem pakar berbasis Certainty Factor, gejala dengan bobot pakar tinggi memiliki pengaruh besar terhadap diagnosis akhir, dan perubahan kecil pada gejala utama dapat memengaruhi hasil diagnosis secara signifikan.
